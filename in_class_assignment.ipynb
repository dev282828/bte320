```python
first_number = float(input("Number 1: "))
operation = input("Operation: ")
second_number = float(input("Number 2: "))


if operation == "+":
  answer = first_number * second_number
  print(answer)
if operation == "-":
  answer = first_number - second_number
  print(answer)
if operation == "*":
  answer = first_number * second_number
  print(answer)
if operation == "/":
  answer = first_number / second_number
  print(answer)

```
