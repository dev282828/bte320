```python
first_number = float(input("Number 1: "))
operation = input("Operation: ")
second_number = float(input("Number 2: "))


if operation == "+":
  answer = first_number * second_number
  print(answer)
elif operation == "-":
  answer = first_number - second_number
  print(answer)
elif operation == "*":
  answer = first_number * second_number
  print(answer)
elif operation == "/":
  answer = first_number / second_number
  print(answer)
```

```python
print("Hour", "Fee")
for hours in range(1,9,1):
  if 2 <= hours <= 6:
    price = 5 + 2.5 * hours
    print(f" {hours}   {price}")
  elif hours > 6:
    print(f" {hours}   20")
  else:
    print(f" {hours}   10")
```
